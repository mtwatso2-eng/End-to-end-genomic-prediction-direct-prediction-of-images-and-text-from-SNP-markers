```python
import os
import cv2
import torch
import numpy as np
from google.colab import drive
import matplotlib.pyplot as plt
drive.mount('/content/drive', force_remount=True)
os.chdir('/content/drive/Shareddrives/Strawberries/Full example')

nReps = 50
randomSeeds = [i for i in range(1, nReps + 1)]
geno = torch.load('fullGeno.pt', map_location=torch.device('cpu')).type(torch.float32)
images = torch.load('fullImages.pt')

trainSplit = 0.8

def to_img(x):
    return np.moveaxis(x.numpy() * 255, 0, -1).astype(np.uint8)
```

```python
for randomSeed in randomSeeds:
  print('Random seed: ' + str(randomSeed))

  _, cloneIndices = torch.unique(geno, dim=0, return_inverse=True)
  clones = torch.unique(cloneIndices)

  nTrain = int(np.round(0.8 * len(clones)))
  nTest = len(clones) - nTrain
  np.random.seed(randomSeed)
  clones = np.random.choice(clones, len(clones), replace = False)
  trainClones, testClones = clones[:nTrain], clones[nTrain:]

  trainKey = np.array([i if i in trainClones else -1 for i in cloneIndices])
  testKey = np.array([i if i in testClones else -1 for i in cloneIndices])

  np.savetxt(str(randomSeed) + '_trainKey.csv', trainKey, delimiter = ',')
  np.savetxt(str(randomSeed) + '_testKey.csv', testKey, delimiter = ',')
```
